# Well-known embedded methods

We compute the stability radius values of these methods and plot thier stability regions, too.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from nodepy import rk
import numpy as np
import math
```

```python
## Fehlberg12

rk.A = np.array([[0,0],[1,0]])
rk.b = np.array([1,0])
rk.bhat = np.array([1/2,1/2])


## Stability plot
rk_pair = rk.ExplicitRungeKuttaPair(rk.A,rk.b,rk.bhat)
# Stability region and save as an eps file
fig1=rk_pair.plot_stability_region(N=1000);
fig1.axes[0].set_title('Absolute stability region')
#fig1.axes[0].set_title('')
#plt.savefig('Fehlberg12.eps', format='eps', dpi=10000)


## Stability radius values
rk_pair_original = rk.ExplicitRungeKuttaPair(rk.A,rk.b,rk.b)

print('\n')
print(rk_pair_original.embedded_method.__num__().real_stability_interval())
print(rk_pair_original.embedded_method.__num__().imaginary_stability_interval())
print(rk_pair_original.embedded_method.__num__().circle_contractivity_radius())
print(rk_pair_original.embedded_method.__num__().absolute_monotonicity_radius())


print('\n')
print(rk_pair.embedded_method.__num__().real_stability_interval())
print(rk_pair.embedded_method.__num__().imaginary_stability_interval())
print(rk_pair.embedded_method.__num__().circle_contractivity_radius())
print(rk_pair.embedded_method.__num__().absolute_monotonicity_radius())
```
