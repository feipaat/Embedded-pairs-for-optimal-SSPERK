# Embedded SSPERK$(10,4)$ pairs

Based on our numerical search we have eight potential embedded pairs. For further details see the paper. Below we compare these pairs and the original method based on different measumerement values from the paper of van der Marel (1990).

```python
%matplotlib inline
import matplotlib.pyplot as plt
from nodepy import rk
import numpy as np
import math
```

```python
ssp104=rk.loadRKM('SSP104')
print ssp104
print ssp104.__num__().real_stability_interval()
print ssp104.__num__().imaginary_stability_interval()
print ssp104.__num__().circle_contractivity_radius()
print ssp104.__num__().absolute_monotonicity_radius()
```
